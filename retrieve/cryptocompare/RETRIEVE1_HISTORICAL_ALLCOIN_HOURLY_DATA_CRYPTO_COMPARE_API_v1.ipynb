# GET INTREVAL HISTORICAL DATA FOR EACH COIN

```python
def flatten(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v, collections.MutableMapping):
            items.extend(flatten(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)
```

```python

iter1 = 0

cont_dict2={}
coin_histlist =  []
error_urllist = [] # records exception urls
coin_df1 = pd.DataFrame()
coin_df2 = pd.DataFrame()
batch = 1

for i in alist:
    # reset dataframe and reset list 
    coin_histlist = []
    coin_df1 = pd.DataFrame()
    coin_df2 = pd.DataFrame()
    print('')
    print('')
    print(iter1)
    #https://min-api.cryptocompare.com/data/v2/histoday?fsym=BTC&tsym=USD&limit=2000
    url_coin = 'https://min-api.cryptocompare.com/data/v2/histohour?fsym='+str(i)+'&tsym=USD&limit=2000' + '&api_key=' + api_key
    print(url_coin)

    # get the main data
    sleep(0.5)
    req_coin = urllib.request.Request(url_coin)
    sleep(0.6)
    r_coin = urllib.request.urlopen(req_coin).read()
    sleep(0.7)
    cont_coin = json.loads(r_coin.decode('utf-8'))
    try:
        # xtract time series from dictionary of dictionaries (json)
        cont_dict2 = cont_coin['Data']['Data']
        
        
        # flattent the data 
        
        for market1 in cont_dict2:
                #print(market)
                flatdict1 = flatten(market1)
                flatdict1['timestamp'] = datetime.datetime.now() # add timestamp to dictionary
                flatdict1['iteration'] = iter1 # add timestamp to dictionary
                coin_histlist.append(flatdict1) # append to list
                
        # save list
        algfile = algdir + i + '_hourlyhist_crypto_compare_api_v1_list' + str(batch)
        picklefile = open(algfile, 'wb')
        pickle.dump(coin_histlist, picklefile)
        picklefile.close() 
        
        # convert to dataframe and save
        coin_df1 = pd.DataFrame(coin_histlist)
        
            
            
        coin_df1['timestamp'] = datetime.datetime.now()
        coin_df1['coin_symbol'] = i


        print('length df is: ', len(coin_df1))

        # save dataframe
        algfile = algdir + i+ '_hourlyhist_crypto_compare_api_v1_df'  + str(batch)
        picklefile = open(algfile, 'wb')
        pickle.dump(coin_df1, picklefile)
        picklefile.close() 

    except:
        print('Exception getting coin')
        error_urllist.append(url_coin)
        #print(cont_coin['Data'])
        # save dataframe
        algfile = algdir + 'error_hourly_crypto_compare_api_v1_list'  + str(batch)
        picklefile = open(algfile, 'wb')
        pickle.dump(error_urllist, picklefile)
        picklefile.close() 
    
    iter1+=1
```
